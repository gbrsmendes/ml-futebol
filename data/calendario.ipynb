```python
import pandas as pd
pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", None)
pd.set_option("display.width", 1000)
pd.options.display.float_format = "{:,.2f}".format

import random
import numpy as np

random.seed(27)
np.random.seed(27)
```

```python
ligas = ["A", "B", "C"]
rodadas = range(1, 39)
partidas = range(1, 11)

idx = pd.MultiIndex.from_product([ligas, rodadas, partidas], names=["serie", "rodada", "partida"])
df_calendario_base = idx.to_frame(index=False)

jogos_de_todas_as_ligas = []

for i, liga in enumerate(ligas):
    
    inicio = (i * 20) + 1
    fim = (i + 1) * 20
    clubes = [f"clube {j:02d}" for j in range(inicio, fim + 1)]
    random.shuffle(clubes)
    
    pivo = clubes[-1]
    clubes_que_giram = clubes[:-1]

    primeiro_turno = []
    
    for r in range(19):
      
        adversario_pivo = clubes_que_giram[0]
        if r % 2 == 0:
            primeiro_turno.append((pivo, adversario_pivo))
        else:
            primeiro_turno.append((adversario_pivo, pivo))
            
        for j in range(1, 10):
            mandante = clubes_que_giram[j]
            visitante = clubes_que_giram[-j]
            
            if r % 2 == 0:
                primeiro_turno.append((mandante, visitante))
            else:
                primeiro_turno.append((visitante, mandante))
                
        clubes_que_giram.insert(0, clubes_que_giram.pop())

    segundo_turno = [(visitante, mandante) for mandante, visitante in primeiro_turno]
    jogos_de_todas_as_ligas.extend(primeiro_turno + segundo_turno)


df_calendario_final = df_calendario_base.copy()
df_calendario_final['mandante'] = [jogo[0] for jogo in jogos_de_todas_as_ligas]
df_calendario_final['visitante'] = [jogo[1] for jogo in jogos_de_todas_as_ligas]

print(
    df_calendario_final
    [df_calendario_final["rodada"].isin([1,20])]
)

df_calendario_final.to_parquet(r"calendario.parquet")
```
